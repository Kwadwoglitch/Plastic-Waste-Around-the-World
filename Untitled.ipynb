```python
import time
from geopy.geocoders import Nominatim
import geopandas as gpd

# List of all country names as provided

# Initialize DataFrame

# Initialize the geolocator
geolocator = Nominatim(user_agent="geoapi")

# Function to get coordinates
def get_coordinates(country):
    try:
        location = geolocator.geocode(country)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except:
        return None, None

# Get coordinates for each country
data[['Latitude', 'Longitude']] = data['Country'].apply(lambda x: pd.Series(get_coordinates(x)))
time.sleep(1)  # Adding delay to avoid rate limits

# Filter out countries without coordinates
data = data.dropna(subset=['Latitude', 'Longitude'])

merged_data.fillna(0)

# Plot the world map and overlay country locations
fig, ax = plt.subplots(figsize=(15, 10))
merged_data.plot(ax=ax, color='lightgray', edgecolor='black')
merged_data['Total_Plastic_Waste_MT'].plot(ax=ax, color='red', markersize=10, label="Country Locations")

# Customize the plot
plt.title("Countries' Locations on the World Map")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend()
plt.show()
```
